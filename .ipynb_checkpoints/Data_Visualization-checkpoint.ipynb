# Genre Popularity

```python
# Query to fetch genres
query_genres = "SELECT DISTINCT genres FROM anime_data"

# Fetch data into a pandas DataFrame
anime = pd.read_sql(query_genres, con=conn)

# Function to count unique genres
def count_genres(genres_list):
    genre_counts = Counter()
    for genres in genres_list:
        # Convert the string representation of a list into an actual list
        genres = ast.literal_eval(genres)
        for genre in genres:
            genre_counts[genre.strip()] += 1
    return genre_counts

# Count genre occurrences
genre_counts = count_genres(anime['genres'])

# Convert the Counter object to a pandas DataFrame
genre_counts_df = pd.DataFrame.from_dict(genre_counts, orient='index').reset_index()
genre_counts_df.columns = ['Genre', 'Count']
genre_counts_df = genre_counts_df.sort_values(by='Count', ascending=False)

# Shorten or abbreviate genre names for the legend
genre_counts_df['Shortened_Genre'] = genre_counts_df['Genre'].apply(lambda x: x[:10] + '...' if len(x) > 10 else x)

# Generate a colormap
num_genres = len(genre_counts)
colors = plt.cm.get_cmap('tab20', num_genres).colors

# Plotting the bar chart for genre popularity
plt.figure(figsize=(14, 8))
bars = plt.bar(genre_counts_df['Shortened_Genre'], genre_counts_df['Count'], color=colors)
plt.xlabel('Genres')
plt.ylabel('Count')
plt.title('Genre Popularity')
plt.xticks(rotation=90)  # Rotate the genre labels for better readability
plt.tight_layout()

# Add color mapping to bars
for bar, color in zip(bars, colors):
    bar.set_color(color)

plt.show()

# Close MySQL connection
conn.close()
```

#  Top Studios by Average Score

```python
# Establish the MySQL connection using SQLAlchemy
engine = create_engine('mysql+mysqlconnector://root:mYsT4nd4rdQu3rYL4ngu4g3@localhost/anime_database')

# Query to fetch top studios by average score
query_top_studios = """
    SELECT studios, AVG(score) AS avg_score
    FROM anime_data
    WHERE studios IS NOT NULL AND score IS NOT NULL
    GROUP BY studios
    ORDER BY avg_score DESC
    LIMIT 10
"""

# Fetch data into a pandas DataFrame
top_studios = pd.read_sql(query_top_studios, con=engine)

# Function to clean and count studios
def count_studios(studios_str):
    studios_list = [studio.strip("[]'") for studio in studios_str.split(', ')]
    return studios_list

# Apply the function to clean studios column
top_studios['studios_cleaned'] = top_studios['studios'].apply(count_studios)

# Plotting the bar chart with average score on x-axis
plt.figure(figsize=(10, 8))
for index, row in top_studios.iterrows():
    for studio in row['studios_cleaned']:
        plt.barh(studio, row['avg_score'], color='green')

plt.xlabel('Average Score')
plt.ylabel('Studio')
plt.title('Top Studios by Average Score')
plt.grid(True)
plt.tight_layout()
plt.show()

# Close SQLAlchemy connection
engine.dispose()
```

# Linear Regression

## Relationship between different anime genres and their associated scores.

```python
import statsmodels.api as sm
from sklearn.preprocessing import MultiLabelBinarizer

engine = create_engine('mysql+mysqlconnector://root:mYsT4nd4rdQu3rYL4ngu4g3@localhost/anime_database')

query = "SELECT genres, score FROM anime_data WHERE genres IS NOT NULL AND score IS NOT NULL"
anime = pd.read_sql(query, con=engine)

anime['genres'] = anime['genres'].apply(lambda x: ast.literal_eval(x))

mlb = MultiLabelBinarizer()
genre_matrix = pd.DataFrame(mlb.fit_transform(anime['genres']), columns=mlb.classes_)

anime_with_genres = pd.concat([anime[['score']], genre_matrix], axis=1)

X = anime_with_genres.drop('score', axis=1)
y = anime_with_genres['score']

# Add a constant
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

coefficients = model.params
coefficients_df = coefficients.reset_index()
coefficients_df.columns = ['Genre', 'Coefficient']

print(coefficients_df)
print("\n",model.summary())

# Visualization
plt.figure(figsize=(12, 6))
sns.barplot(x='Coefficient', y='Genre', data=coefficients_df.sort_values(by='Coefficient', ascending=False))

# Add title and labels
plt.title('Genre Coefficients from OLS Regression Model', fontsize=16)
plt.xlabel('Coefficient', fontsize=14)
plt.ylabel('Genre', fontsize=14)

# Show the plot
plt.show()

engine.dispose()
```

## Relationship between Viewer Favorites and Anime Scores

```python
# Establish database connection
engine = create_engine('mysql+mysqlconnector://root:mYsT4nd4rdQu3rYL4ngu4g3@localhost/anime_database')

# Query data (adjust query as per your database schema)
query = "SELECT score, favorites FROM anime_data WHERE score IS NOT NULL AND favorites IS NOT NULL"
anime = pd.read_sql(query, con=engine)

# Close database connection
engine.dispose()

# Perform linear regression
X = anime[['favorites']]
y = anime['score']

# Add a constant for intercept
X = sm.add_constant(X)

# Fit the model
model = sm.OLS(y, X).fit()

# Print coefficients and summary
print(model.params)
print("\n", model.summary())

# Visualization 
plt.figure(figsize=(10, 6))
sns.regplot(x='favorites', y='score', data=anime, scatter_kws={'alpha':0.5})
plt.title('Relationship between Viewer Favorites and Anime Scores')
plt.xlabel('Viewer Favorites')
plt.ylabel('Anime Score')
plt.ylim(0, 10)  # Set y-axis limit to 0-10
plt.grid(True)
plt.show()
```

```python
# Establish database connection
engine = create_engine('mysql+mysqlconnector://root:mYsT4nd4rdQu3rYL4ngu4g3@localhost/anime_database')

# Query data (adjust query as per your database schema)
query = """
    SELECT score, CAST(start_season AS CHAR) AS start_season
    FROM anime_data 
    WHERE score IS NOT NULL AND start_season IS NOT NULL
"""
anime = pd.read_sql(query, con=engine)

# Close database connection
engine.dispose()

# Create dummy variables for seasons
season_dummies = pd.get_dummies(anime['start_season'], prefix='season')

# Concatenate dummies with original dataframe
anime_with_dummies = pd.concat([anime['score'], season_dummies], axis=1)

# Prepare data for regression
X = anime_with_dummies.drop('score', axis=1)  # Independent variables (season dummies)
y = anime_with_dummies['score']  # Dependent variable (score)

# Convert bool columns in X to int64
X = X.astype('int64')

# Add a constant for intercept
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Print coefficients and summary
print("Coefficients:\n", model.params)
print("\n", model.summary())

# Visualization of coefficients
plt.figure(figsize=(10, 6))
sns.barplot(x=season_dummies.columns, y=model.params[1:])
plt.title('Seasonal Effects on Anime Scores')
plt.xlabel('Season')
plt.ylabel('Effect on Score')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
